```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import sklearn.datasets as datasets #导入内置数据
```

决策树分类

```python
iris = datasets.load_iris()
```

```python
data = iris.data
target = iris.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(data,target,train_size=0.8)
```

```python
tree = DecisionTreeClassifier()
tree.fit(x_train,y_train)
```

```python
tree.score(x_test,y_test)
print(tree.tree_)
```

```python
lg = LogisticRegression()
lg.fit(x_train,y_train)
```

```python
lg.score(x_test,y_test)
```

决策树回归

```python
x = 200*np.random.rand(500,1)-100
x_train = np.sort(x,axis=0)
y_train = np.pi*np.array([np.sin(x_train).ravel(),np.cos(x_train).ravel()])
y_train = y_train.T
y_train[::25] +=np.random.rand(20,2)
```

```python
plt.scatter(y_train[:,0],y_train[:,1])
```

```python
tree1 = DecisionTreeRegressor(max_depth=5)
tree2 = DecisionTreeRegressor(max_depth=10)
tree3 = DecisionTreeRegressor(max_depth=50)
tree1.fit(x_train,y_train)
tree2.fit(x_train,y_train)
tree3.fit(x_train,y_train)
```

```python
x_test = np.arange(-100,100,0.1).reshape((-1,1))
```

```python
y_predict1 = tree1.predict(x_test)
y_predict2 = tree2.predict(x_test)
y_predict3 = tree3.predict(x_test)
```

```python
plt.figure(figsize=(4,12))
plt.subplot(411)
plt.scatter(y_train[:,0],y_train[:,1])
plt.subplot(412)
plt.scatter(y_predict1[:,0],y_predict1[:,1])
plt.subplot(413)
plt.scatter(y_predict2[:,0],y_predict2[:,1])
plt.subplot(414)
plt.scatter(y_predict3[:,0],y_predict3[:,1])
```
