```python
import numpy as np
import sklearn.datasets
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier #分类
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap #绘图引用模块
import sklearn.datasets as datasets
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
x_train,x_test,y_train,y_test = train_test_split(iris.data,iris.target,train_size=0.3)
```

```python
knn = KNeighborsClassifier(10,algorithm='kd_tree')
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
cmap = ListedColormap(["#FF0000", "#00FF00", "#0000FF"])
```

```python
plt.scatter(iris.data[:,2],iris.data[:,3],c=iris.target,cmap=cmap)
```

```python
x = iris.data[:,:2]
y = iris.target
#设置网格格式参数
xmin,xmax = x[:,0].min()-1,x[:,0].max()-1
ymin,ymax = x[:,1].min()-1,x[:,1].max()-1
h = 0.02
#生成网格
xx,yy=np.meshgrid(np.arange(xmin,xmax,h),np.arange(ymin,ymax,h))
#设置颜色版
cmap_light = ListedColormap(["#FFAAAA", "#AAFFAA", "#AAAAFF"])
cmap_bold = ListedColormap(["#FF0000", "#00FF00", "#0000FF"])
#设置KNN
newknn = KNeighborsClassifier(15)
newknn.fit(x,y)
z = newknn.predict(np.c_[xx.ravel(),yy.ravel()])
z = z.reshape(xx.shape)
#显示背景
plt.pcolormesh(xx,yy,z,cmap=cmap_light)
#显示点 
plt.scatter(x[:,0],x[:,1],c=y,cmap=cmap_bold)

plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.title("Classifier")
```
