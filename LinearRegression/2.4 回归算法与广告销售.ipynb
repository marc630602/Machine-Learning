```python
x_train,x_test,y_train,y_test = train_test_split(data.iloc[:,:-1],data.iloc[:,-1],train_size=0.8,random_state=1 )
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
lr.score(x_test,y_test)
```

```python
plt.figure(facecolor='w',figsize=(15,15))
x = np.arange(0,300,2)
y = lr.coef_[0]*x + lr.intercept_
plt.subplot(311)
plt.plot(data['TV'],data['sales'],'ro',label='TV')
plt.plot(x,y)
plt.title("TV")
plt.xlabel("广告花销")
plt.ylabel("广告收入")
plt.legend()
plt.grid()

x = np.arange(0,50,2)
y = lr.coef_[1]*x + lr.intercept_
plt.subplot(312)
plt.plot(data['radio'],data['sales'],'g^',label='Radio')
plt.plot(x,y)
plt.title('Radio')
plt.xlabel("广告花销")
plt.ylabel("广告收入")
plt.legend()
plt.grid()

x = np.arange(0,100,2)
y = lr.coef_[2]*x + lr.intercept_
plt.subplot(313)
plt.plot(data['newspaper'],data['sales'],'bv',label='Newspaper')
plt.plot(x,y)
plt.title('Newspaper')
plt.xlabel("广告花销")
plt.ylabel("广告收入")
plt.legend()
plt.grid()
```
