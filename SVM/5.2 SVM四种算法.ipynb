```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.svm import SVR
import matplotlib.pyplot as plt
import sklearn.datasets as datasets 
from matplotlib.colors import ListedColormap
```

```python
iris= datasets.load_iris()
x = iris.data[:,:2]
y = iris.target
x1_min,x1_max = x[:,0].min(),x[:,0].max()
x2_min,x2_max = x[:,1].min(),x[:,1].max()
h=0.1
x1 = np.arange(x1_min,x1_max,h)
x2 = np.arange(x2_min,x2_max,h)
xx1,xx2 = np.meshgrid(x1,x2)
x_test = np.c_[xx1.ravel(),xx2.ravel()]
```

```python
estimator = [SVC(kernel='linear'),
             SVC(kernel='rbf'),
             SVC(kernel='poly'),
             SVC(kernel='sigmoid')]
```

```python
titles = ['linear_svc','rbf_svc','poly_svc','sigmoid_svc']
plt.figure(figsize=(12,9))
color = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
for i,svc in enumerate(estimator):
    svc.fit(x,y)
    y_predict = svc.predict(x_test)
    z = y_predict.reshape(xx2.shape)
    plt.subplot(2,2,i+1)
    plt.title(titles[i])
    plt.scatter(x[:,0],x[:,1],c=y,cmap=color)
    plt.contourf(xx1,xx2,z,cmap='cool',alpha=0.5)  #绘制轮廓
```

```python
a = SVC(kernel='linear')
a.fit(x,y)
z=a.predict(x_test)
plt.subplot(2,2,1)
plt.title(titles[3])
plt.scatter(x[:,0],x[:,1],c=y,cmap=color)
#plt.contourf(xx1,xx2,z,cmap='cool',alpha=0.5)
```
